# Investigate missing data
### A look at data excluded from original analysis



## Read in the SAS dataset and create DBD and DCD approach cohorts to match those used in previous analysis cohort 

```python
#import libraries
import pandas as pd
from scipy.stats import f_oneway
```

```python
#Read in dataset with all rows included
df = pd.read_sas("Data/alldata3.sas7bdat")

#6931 DBD apps
dbd_apps_all = df[(df["eli_DBD"]==1)&(df["FAMILY_APPROACHED"]==2)]
print(str(dbd_apps_all.shape[0]) + " DBD approaches")

#6060 DBD apps to match cohort in paper
dbd_apps = dbd_apps_all[(dbd_apps_all["eth_grp"]!=5)&(dbd_apps_all["FORMAL_APR_WHEN"]!=4)&(dbd_apps_all["donation_mentioned"]!=-1)
                    &(dbd_apps_all["FAMILY_WITNESS_BSDT"]!=9)&(dbd_apps_all["GENDER"]!=9)]
print(str(dbd_apps.shape[0]) + " DBD approaches with missing and unknown data removed")

#871 DBD apps to match cohort in paper
dbd_apps_removed = dbd_apps_all[(dbd_apps_all["eth_grp"]==5)|(dbd_apps_all["FORMAL_APR_WHEN"]==4)|(dbd_apps_all["donation_mentioned"]==-1)
                    |(dbd_apps_all["FAMILY_WITNESS_BSDT"]==9)|(dbd_apps_all["GENDER"]==9)]
print(str(dbd_apps_removed.shape[0]) + " DBD approaches with missing or unknown data\n")
     
#560 DCD apps removed
dcd_apps_all = df[(df["eli_DCD"]==1)&(df["FAMILY_APPROACHED"]==2)]
print(str(dcd_apps_all.shape[0]) + " DCD approaches")

#9405 DCD apps to match cohort in paper
dcd_apps = dcd_apps_all[(dcd_apps_all["GENDER"]!=9)&(dcd_apps_all["cod_neuro"].notna())&(dcd_apps_all["eth_grp"]!=5)&(dcd_apps_all["donation_mentioned"]!=-1)&
                    (~dcd_apps_all["DTC_WD_TRTMENT_PRESENT"].isin([8,9]))]
print(str(dcd_apps.shape[0]) + " DCD approaches with missing and unknown data removed")

#9405 DCD apps to match cohort in paper
dcd_apps_removed = dcd_apps_all[(dcd_apps_all["GENDER"]==9)|(dcd_apps_all["cod_neuro"].isna())|(dcd_apps_all["eth_grp"]==5)|(dcd_apps_all["donation_mentioned"]==-1)|
                    (dcd_apps_all["DTC_WD_TRTMENT_PRESENT"].isin([8,9]))]
print(str(dcd_apps_removed.shape[0]) + " DCD approaches with missing or unknown data\n")


dbd_cons = dbd_apps[dbd_apps["FAMILY_CONSENT"]==2]
dcd_cons = dcd_apps[dcd_apps["FAMILY_CONSENT"]==2]
print(str(dbd_cons.shape[0]) + " DBD consents with missing and unknown data removed")
print(str(dcd_cons.shape[0]) + " DCD consents with missing and unknown data removed")

dbd_cons_removed = dbd_apps_removed[dbd_apps_removed["FAMILY_CONSENT"]==2]
dcd_cons_removed = dcd_apps_removed[dcd_apps_removed["FAMILY_CONSENT"]==2]
print(str(dbd_cons_removed.shape[0]) + " DBD consents with missing or unknown data")
print(str(dcd_cons_removed.shape[0]) + " DCD consents with missing or unknown data\n")

consent_rate_dbd_removed = len(dbd_cons_removed.index)/len(dbd_apps_removed.index)
print("DBD consent rate for removed data: " + str(consent_rate_dbd_removed))
consent_rate_dbd = len(dbd_cons.index)/len(dbd_apps.index)
print("DBD consent rate: " + str(consent_rate_dbd))

consent_rate_dcd_removed = len(dcd_cons_removed.index)/len(dcd_apps_removed.index)
print("\nDCD consent rate for removed data: " + str(consent_rate_dcd_removed))
consent_rate_dcd = len(dcd_cons.index)/len(dcd_apps.index)
print("DCD consent rate: " + str(consent_rate_dcd))
```

```python
print(dbd_apps["AGE_YEARS"].describe()["mean"])
print(dbd_apps_removed["AGE_YEARS"].describe()["mean"])

print(dcd_apps["AGE_YEARS"].describe()["mean"])
print(dcd_apps_removed["AGE_YEARS"].describe()["mean"])

print(dbd_cons["AGE_YEARS"].describe()["mean"])
print(dbd_cons_removed["AGE_YEARS"].describe()["mean"])

print(dcd_cons["AGE_YEARS"].describe()["mean"])
print(dcd_cons_removed["AGE_YEARS"].describe()["mean"])

# No significant difference in mean age between groups used and 
print("\np=" + str(f_oneway(dbd_apps["AGE_YEARS"],dbd_apps_removed["AGE_YEARS"])[1]))
print("\np=" + str(f_oneway(dcd_apps["AGE_YEARS"],dcd_apps_removed["AGE_YEARS"])[1]))
```
